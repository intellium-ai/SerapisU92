```python
import numpy as np
import pandas as pd
```

```python
raw_df = pd.read_csv('data/energetic_original.csv')
raw_df
```

```python
from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Descriptors

for i, row in raw_df.iterrows():
    molecule = Chem.MolFromSmiles(row.SMILES)
    raw_df.at[i, 'Number of Atoms'] = molecule.GetNumAtoms()
    raw_df.at[i, 'Exact Molecular Weight'] = Descriptors.ExactMolWt(molecule)
    raw_df.at[i, 'Average Molecular Weight'] = Descriptors.MolWt(molecule)
    raw_df.at[i, 'Heavy Atoms Molecular Weight'] = Descriptors.HeavyAtomMolWt(molecule)

    raw_df.at[i, 'Number of Radical Electrons'] = Descriptors.NumRadicalElectrons(molecule)
    raw_df.at[i, 'Number of Valence Electrons'] = Descriptors.NumValenceElectrons(molecule)

    raw_df.at[i, 'Fp Density Morgan 1'] = Descriptors.FpDensityMorgan1(molecule)
    raw_df.at[i, 'Fp Density Morgan 2'] = Descriptors.FpDensityMorgan2(molecule)
    raw_df.at[i, 'Fp Density Morgan 3'] = Descriptors.FpDensityMorgan3(molecule)
    
    raw_df.at[i, 'Max Absolute Partial Charge'] = Descriptors.MaxAbsPartialCharge(molecule)
    raw_df.at[i, 'Max Partial Charge'] = Descriptors.MaxPartialCharge(molecule)
    raw_df.at[i, 'Min Absolute Partial Charge'] = Descriptors.MinAbsPartialCharge(molecule)
    raw_df.at[i, 'Min Partial Charge'] = Descriptors.MinPartialCharge(molecule)

    raw_df.at[i, 'BalabanJ'] = Descriptors.BalabanJ(molecule)
    raw_df.at[i, 'BertzCT'] = Descriptors.BertzCT(molecule)
    raw_df.at[i, 'Chi0'] = Descriptors.Chi0(molecule)
    raw_df.at[i, 'Chi1'] = Descriptors.Chi1(molecule)

    raw_df.at[i, 'EState_VSA1'] = Descriptors.EState_VSA1(molecule)
    raw_df.at[i, 'EState_VSA2'] = Descriptors.EState_VSA2(molecule)
    raw_df.at[i, 'EState_VSA3'] = Descriptors.EState_VSA3(molecule)
    raw_df.at[i, 'EState_VSA4'] = Descriptors.EState_VSA4(molecule)
    raw_df.at[i, 'EState_VSA5'] = Descriptors.EState_VSA5(molecule)
    raw_df.at[i, 'EState_VSA6'] = Descriptors.EState_VSA6(molecule)
    raw_df.at[i, 'EState_VSA7'] = Descriptors.EState_VSA7(molecule)
    raw_df.at[i, 'EState_VSA8'] = Descriptors.EState_VSA8(molecule)
    raw_df.at[i, 'EState_VSA9'] = Descriptors.EState_VSA9(molecule)
    raw_df.at[i, 'EState_VSA10'] = Descriptors.EState_VSA10(molecule)
    raw_df.at[i, 'EState_VSA11'] = Descriptors.EState_VSA11(molecule)

    raw_df.at[i, 'VSA_EState1'] = Descriptors.VSA_EState1(molecule)
    raw_df.at[i, 'VSA_EState2'] = Descriptors.VSA_EState2(molecule)
    raw_df.at[i, 'VSA_EState3'] = Descriptors.VSA_EState3(molecule)
    raw_df.at[i, 'VSA_EState4'] = Descriptors.VSA_EState4(molecule)
    raw_df.at[i, 'VSA_EState5'] = Descriptors.VSA_EState5(molecule)
    raw_df.at[i, 'VSA_EState6'] = Descriptors.VSA_EState6(molecule)
    raw_df.at[i, 'VSA_EState7'] = Descriptors.VSA_EState7(molecule)
    raw_df.at[i, 'VSA_EState8'] = Descriptors.VSA_EState8(molecule)
    raw_df.at[i, 'VSA_EState9'] = Descriptors.VSA_EState9(molecule)
    raw_df.at[i, 'VSA_EState10'] = Descriptors.VSA_EState10(molecule)
```
